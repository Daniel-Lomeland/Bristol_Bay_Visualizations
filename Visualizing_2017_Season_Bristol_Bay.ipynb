```python
import pandas as pd
import numpy as np
import datetime 

# Ignore Warnings
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('Bristol_Bay_Daily_Catch_2017.csv')

df.rename(columns={'District Name': 'District',
                   'Chinook Count' :'Chinook',
                   'Sockeye Count':'Sockeye',
                   'Coho Count':'Coho',
                   'Pink Count':'Pink',
                   'Chum Count':'Chum',
                   'Total Count':'Daily_Total'
                   
                  }, inplace=True)

df['District']= df['District'].str.replace('321 - Ugashik','Ugashik')
df['District']= df['District'].str.replace('322 - Egegik','Egegik')
df['District']= df['District'].str.replace('324 - Naknek-Kvichak','Naknek-Kvichak')
df['District']= df['District'].str.replace('325 - Nushagak','Nushagak')
df['District']= df['District'].str.replace('326 - Togiak','Togiak')
df.head(1)
```

```python
df1= df[['District','Chinook','Sockeye','Coho','Pink','Chum']]
df1.head()
```

```python
district_df = df1.groupby(['District'], as_index=False).sum()
district_df['Target']= district_df['District']

district_df['Target']= district_df['Target'].str.replace('Ugashik','5')
district_df['Target']= district_df['Target'].str.replace('Egegik','6')
district_df['Target']= district_df['Target'].str.replace('Naknek-Kvichak','7')
district_df['Target']= district_df['Target'].str.replace('Nushagak','8')
district_df['Target']= district_df['Target'].str.replace('Togiak','9')
district_df['Target']= district_df['Target'].astype(int)
district_df
```

```python

sockeye_df = district_df[['District','Sockeye']]
sockeye_df['Name']= sockeye_df['District']
sockeye_df['Source']=0
sockeye_df.rename(columns={'District':'Target',
                           'Sockeye':'Value'
                          }, inplace=True)
sockeye_df
```

```python
sockeye_df['Target']= sockeye_df['Target'].str.replace('Ugashik','5')
sockeye_df['Target']= sockeye_df['Target'].str.replace('Egegik','6')
sockeye_df['Target']= sockeye_df['Target'].str.replace('Naknek-Kvichak','7')
sockeye_df['Target']= sockeye_df['Target'].str.replace('Nushagak','8')
sockeye_df['Target']= sockeye_df['Target'].str.replace('Togiak','9')
sockeye_df['Target']= sockeye_df['Target'].astype(int)
```

```python
sockeye_df['Value']= sockeye_df['Value'].astype(int)
sockeye_df
```

```python
import plotly.plotly as py

data = dict(
    type='sankey',
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(
        color = "black",
        width = 0.5
      ),
      label = ["Sockeye","Ugshik","Egegik"," Naknek-Kvichak","Nushagak","Togiak" ],
      color = ["blue", "red", "yellow", "green", "orange", "black"]
    ),
    link = dict(
      source = sockeye_df['Source'],
      target = sockeye_df['Target'],
      value = sockeye_df['Value']
  ))

layout =  dict(
    title = "Basic Sankey Diagram",
    font = dict(
      size = 10
    )
)

fig = dict(data=[data], layout=layout)
py.iplot(fig, validate=False)
```

```python
# Group results by date
date_df = df.groupby(['Catch Date','District'], as_index=False).sum()

# Convert Catch Date to datetime format
date_df['Catch Date'] = pd.to_datetime(date_df['Catch Date'])

# Create Week number column
date_df['Week']= date_df['Catch Date'].dt.week

# Filter out insignificant weeks
date_df = date_df.query('25 <= Week <= 31')

# Group by week
week_df = date_df.groupby(['Week','District'], as_index=False).sum()
week_df['Name']= week_df['District']
week_df.head()

# add District number
week_df['District']= week_df['District'].str.replace('Ugashik','5')
week_df['District']= week_df['District'].str.replace('Egegik','6')
week_df['District']= week_df['District'].str.replace('Naknek-Kvichak','7')
week_df['District']= week_df['District'].str.replace('Nushagak','8')
week_df['District']= week_df['District'].str.replace('Togiak','9')
week_df['Target']= week_df['District'].astype(int)
week_df.head()
```

```python
ws_df= week_df[["Name","Week","District","Sockeye"]]
ws_df['Color']= "#add6b4"
ws_df.head()
```

```python
data = dict(
    type='sankey',
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(
        color = "black",
        width = 0.5
      ),
      label = ws_df['Name'],
      color = ws_df['Color']
    ),
    link = dict(
      source = ws_df['District'],
      target = ws_df['Target'],
      value = ws_df['Value']
  ))

layout =  dict(
    title = "Basic Sankey Diagram",
    font = dict(
      size = 10
    )
)

fig = dict(data=[data], layout=layout)
py.iplot(fig, validate=False)
```
